```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### Load dataset

```python
df=sns.load_dataset('tips')
df
#quantative dataset -> total_bills,tip
#qualitative/categorical data -> sex,smoker,day,time,size
```

### Displot/Histplot

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
#kde->kernel distribution estimator(-ve, +ve, symmetric distribution)
#Data is +vely skewed
```

```python
#Applying log transformation
df['total_bill_log']=np.log(df['total_bill'])
df['tip']=np.log(df['tip'])
plt.subplot(1,2,1)
sns.histplot(df['total_bill_log'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

### Jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

### Pairplot

```python
sns.pairplot(df,hue='size')
```

### Rugplot

```python
sns.rugplot(df)
```
