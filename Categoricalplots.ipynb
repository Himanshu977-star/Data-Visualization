```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### load dataset

```python
df=sns.load_dataset('tips')
df
```

### Countplot


```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

### barplot

```python
#x-> categorical
#y-> quantative
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.sum) #by default mean
```

### boxplot

```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

### violinplot

```python
sns.violinplot(x='tip',y='day',data=df,palette='rainbow')
#Here sat-> has more spreading data means more outliers
#while fri-> has less spreading data mean less or no outliers
```

### stripplot

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

### swarmplot


```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

Here violinplot and swarmplot is almost same as it both make violin shape where data is more but it is not usable with large dataset

```python
sns.violinplot(x='day',y='total_bill',data=df)
sns.swarmplot(x='day',y='total_bill',data=df)
```
