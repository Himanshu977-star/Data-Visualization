```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Matplotlib Basic

```python
X=np.linspace(0,5,11)
X
```

```python
y=X**2
y
```

```python
plt.plot(X,y)
plt.title("Basic Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

## Subplot in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(y**3,y)
plt.subplot(2,2,2)
plt.plot(X,y)
plt.subplot(2,2,3)
plt.plot(y,X)
plt.subplot(2,2,4)
plt.plot(X**2,X**3)
```

## Object Oriented Way

```python
fig=plt.figure(figsize=(5,5),dpi=100) #figsize(about width and height) & dpi(dot per inch)
#fig.add_axes([left, bottom, width, height])
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(X,y)
axis2=fig.add_axes([0.3,0.6,0.3,0.3])
axis2.plot(y,X)
```

## Types of Plot

```python
plt.scatter(X,y)
```

```python
from random import sample
data=sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(X)
```

## Saving plot

```python
fig=plt.figure()
axis=fig.add_axes([0.1,0.1,1,1])
axis.plot(X,y)
fig.savefig("plot.png")
```
